[**Pearson Correlation Coefficients between variables in Dataset Using Heatmap**](https:)

```python
corr_df=df_tracks.drop(["key","mode","explicit"],axis=1).corr(method="pearson")
plt.figure(figsize=(15,6))
heatmap=sns.heatmap(corr_df,annot=True,fmt=".1g", vmin=-1, vmax=1, center=0, cmap="inferno", linewidths=1, linecolor="red")
heatmap.set_title("correlation heatmap between variable")
heatmap.set_xticklabels(heatmap.get_xticklabels(),rotation=90)
```

[**Creating Sample Dataset for Regression Plot**](https://)

```python
sample_df = df_tracks.sample(int(0.004*len(df_tracks)))
print(len(sample_df))
```

[**Correlation between Energy and loudness**](https://)

```python
plt.figure(figsize=(10,5))
sns.regplot(data = sample_df, y="loudness", x="energy", color="g").set(title="Energy v/s Loudness Correlation")
```

[**Correlation Between Popularity and Acousticness**](https://)

```python
plt.figure(figsize=(10,5))
sns.regplot(data = sample_df, y="popularity", x="acousticness", color="purple").set(title="Popularity v/s Acousticness Correlation")
```

[**Changing column name duration_ms to duration**](https://)

```python
df_tracks["duration"]= df_tracks["duration_ms"].apply(lambda x: round(x/1000))
df_tracks.drop('duration_ms',inplace=True, axis=1)
```

```python
df_tracks.head()
```

[**Using Scatter Plot to find Danceability based on Duration**](https://)

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x='duration', y='danceability', data=df_tracks)
plt.title('Scatter Plot: Duration vs Danceability')
plt.show()
```

[**Correaltion between Valence(positiveness) and Tempo**](https://)

```python
plt.figure(figsize=(10,5))
sns.regplot(data = sample_df, y="valence", x="tempo", color="orange").set(title="Valence v/s Tempo Correlation")
```

[**Using Boxplot to find Distrubution of duration**](https://)

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x=df_tracks['duration'])
plt.title('Distribution of Duration')
plt.show()
```

[**Using Countplot to find Explicit and Non-Explicit Tracks**](https://)

```python
plt.figure(figsize=(8, 5))
sns.countplot(x=df_tracks['explicit'])
plt.title('Count of Explicit vs Non-Explicit Tracks')
plt.xlabel('Explicit')
plt.ylabel('Count')
plt.show()
```

[**Extracting Year from Release date**](https://)

```python
df_tracks['release_date'] = pd.to_datetime(df_tracks['release_date'])
```

```python
df_tracks['year'] = df_tracks['release_date'].dt.year
```

```python
year=df_tracks['year']
```

[**Using Lineplot to find Average Duration of Songs Over the Years**](https://)

```python
total_dr=df_tracks.duration
sns.set_style(style="whitegrid")
fig_dims = (10,5)
fig, ax = plt.subplots(figsize=fig_dims)
fig = sns.lineplot(x=year,y=total_dr,ax=ax).set(title="Year v/s Duration")
plt.xticks(rotation=60)
```
